```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
from selenium import webdriver
import time
from tqdm import tqdm
```

```python
link = 'https://stock-pictures.netlify.app/'
driver = webdriver.Chrome()
```

```python
driver.get(link)
```

```python
for i in tqdm(range(0,500000,1000)):
    driver.execute_script('window.scrollTo(0,' + str(i) + ')')
    time.sleep(.5)
```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup.text
```

```python
data = []
for sp in soup.find_all('div',class_ ='container'):
    links    = sp.find('img').get('src')
    tags     = list(set(sp.find('div',class_ = 'tags').text[7:].replace(',','').split(' ')))
    likes    = int(sp.find('div', class_ = 'likes-comments').find_all('span')[0].text[:-6])   
    comments = int(sp.find('div', class_ = 'likes-comments').find_all('span')[1].text[:-9])
    data.append([links,' '.join(tags),likes,comments])
```

```python
df = pd.DataFrame(data, columns = ['Links', 'Tags', 'Likes' , 'Comments'])
df.head()
```

```python
df.to_csv('data.csv', index = False)
```

#### Defining Image Downloader Function

```python
def download(link, img_path ):

    res = requests.get(link)         # Sending the Request
    fd = open(img_path,'wb')         # Getting Image in WB Format
    fd.write(res.content)            # Saving the Image
    fd.close()
```

### Downloading all images

```python
paths   = []
img_ids = []

ref = 100000
i   = 1

for link in tqdm(df['Links']):
    
    img_id = 'A' + str(i + ref)[1:]          # Creating Image ID
    i += 1
    
    img_path = 'Imgs/' + img_id + '.jpg'    # Creating Image Path
    
    download(link, img_path)                # Downloading the Image
    
    paths.append(img_path)
    img_ids.append(img_id)
```

#### Saving Ids and Paths in DataFrame

```python
df['img_id']   = img_ids
df['img_path'] = paths

df.head()
```

```python
df.to_csv('final.csv', index = False)
```
